working with seikimge

```python
from skimage.io import imread
from skimage.transform import resize, rescale, downscale_local_mean
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
!ls -l
```

```python
img = imread("/content/medicine-doctor-stethoscope-hand-touching-600w-1438221293.jpg", as_gray=True)
```

```python
img.shape, type(img)
```

```python
img_rescaled = rescale(img,1./4., anti_aliasing=True)
```

```python
img_rescaled.shape
```

```python
img_resized = resize(img_rescaled, (200,200))
```

```python
plt.imshow(downscale_local_mean(img,(10,10)))

```

image filters edge detectors

```python
from skimage.filters import roberts,sobel,prewitt,scharr
```

```python
fig , axes = plt.subplots(2,2,figsize=(8,8))
ax = axes.ravel()

ax[0].imshow(roberts(img))

ax[1].imshow(sobel(img))

ax[2].imshow(prewitt(img))

ax[3].imshow(scharr(img))


```

```python
from skimage.feature import canny
plt.imshow(canny(img, sigma=2))
```
